```python
import numpy as np
import matplotlib.pyplot as plt

# Part A
# 1.I
nI = np.arange(0, 10)
d = 1 * (nI == 3) # delta[n-3]

plt.figure()
plt.stem(nI, d)
plt.xticks(np.arange(min(nI), max(nI)+1, 1))
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title('1.I) x[n] = delta[n-3]')
plt.show

# 1.II
nII = np.arange(-5, 10)
u = (np.mod(nII,1)==0) * 1 * (nII >= -1) # u[n+1]

plt.figure()
plt.stem(nII, u)
plt.xticks(np.arange(min(nII), max(nII)+1, 1))
plt.title('1.II) x[n] = u[n+1]')
plt.xlabel("n")
plt.ylabel("x[n]")
plt.show

# xn anon function (ex. x[n-3] = x(n, 3))
x = lambda n, pos_shift : np.cos(np.pi * (n-pos_shift) / 5) * 1 * ((n-pos_shift) >= 0) * (np.mod(n,1)==0)

# 1.III
nIII = np.arange(-5, 15)
xn = x(nIII, 0)

plt.figure()
plt.stem(nIII, xn)
plt.xticks(np.arange(min(nIII), max(nIII)+1, 1))
plt.title('1.III) x[n] = cos(n*pi/5)u[n+1]')
plt.xlabel("n")
plt.ylabel("x[n]")
plt.show

# 1.IV
nIV = np.arange(-5, 15)
x1 = x(nIV, 3)

plt.figure()
plt.stem(nIV, x1)
plt.xticks(np.arange(min(nIV), max(nIV)+1, 1))
plt.title('1.IV) x1[n] = x[n-3]')
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.show

# 1.V
nV = np.arange(-15, 5)
x2 = x(-nV, 0)

plt.figure()
plt.stem(nV, x2)
plt.xticks(np.arange(min(nV), max(nV)+1, 1))
plt.title('1.V) x2[n] = x[-n]')
plt.xlabel("n")
plt.ylabel("x2[n]")
plt.show
```

```python
# Part B
# 1) I = 8, D = 2
I = lambda I : 1.0 + ((I/100.0) + 0.01)
interest = I(8)
D = lambda D : (D + 1) * 1000
default_bal = D(2)

# y[n] = balance, y[n-1] = balance after last interest return, x[n] = deposit at current time
# y[n] = (y[n-1] * interest) + x[n]

# interest, starting balance, deposit, deposit factor (100 = 0, 100n = 1), months
def calc_bal(I, D, x, n, months) : 
    y_array = np.zeros(months+1)
    y_array[0] = D;
    
    for i in range(1, months+1) : # 1 to 12
        y_array[i] = (y_array[i-1] * interest) + (x * (i ** n))
        print("Month", i,  f": ${y_array[i]:.2f}")

    t = np.arange(0, 13)
    plt.figure()
    plt.stem(t, y_array)
    plt.xticks(np.arange(min(t), max(t)+1, 1))
    plt.title(f"(y[n-1] * interest) + {x}n")
    plt.ylabel("total balance ($)")
    plt.xlabel("month")
    plt.show


deposit = 100
months = 12
calc_bal(interest, default_bal, 0, 0, months)
print("\n")
calc_bal(interest, default_bal, deposit, 1, months)
```
